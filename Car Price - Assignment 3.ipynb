### Build Linear Regression model
My approach:-
1. Model 1 -> Get top few features from RFE. I will run RFE on different number of selected features till all p values are less than 0.05
2. After that we will build different models manually by checking p_values and VIF of independent variables

```python
#importing all required libraries
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
```

### Model 1 - RFE features selection
I want to come to a model which has all p_values less than 0.05 and as soon as we reach that stage, we will break out of the for loop

```python
#This for loop tries to select top 'n' features which have their p_values < 0.05
#I am using sklearn's RFE to select the top n features and then getting the max p_value from the statsmodel
#after fitting. 
for i in range(X_train.shape[1]):
    lr_m1 = LinearRegression()
    lr_m1.fit(X_train, y_train)
    rfe_m1 = RFE(lr_m1, X_train.shape[1]-i)
    rfe_m1 = rfe_m1.fit(X_train, y_train)
    X_train_rfe = X_train.drop(list(X_train.columns[~rfe_m1.support_]), axis=1)
    X_lm = sm.add_constant(X_train_rfe)
    lr = sm.OLS(y_train, X_lm).fit()
    if lr.pvalues.max()<0.05:
        break
    
```

```python
#checking shape of our selected features from RFE. So we have 19 features in our Model 1
X_train_rfe.shape
```

```python
#defining a check_model function which will check the summary from statsmodel and VIF of each model we built 
#We won't have to write the same code again n again. 
def check_model(X,y):
    # Add a constant
    X_lm = sm.add_constant(X)
    # Create a first fitted model
    lr = sm.OLS(y, X_lm).fit()
    # Print a summary of the linear regression model obtained
    print(lr.summary())
    # Create a dataframe that will contain the names of all the feature variables and their respective VIFs
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    print('************VIF************')
    print(vif)
```

```python
check_model(X_train_rfe, y_train)
```

### Model 2

```python
X_train_model2 = X_train_rfe.drop('carwidth', axis=1)
check_model(X_train_model2, y_train)
```

```python
lm_final= sm.OLS(y_train,sm.add_constant(X_train_model5)).fit()
```

```python
lm_final.summary()
```

#### So we have our final model with 7 independent variables. All these variables are significant as they low p values(<0.05) and low VIF (<5) which means they are not dependent on each other. Our final car price indicators and their corresponding cofficients are:-
- horsepower  0.7923
- carbody_hatchback  -0.0470
- enginelocation_rear  0.2581	
- enginetype_l 	 0.0889
- CompanyName_bmw  0.2883	
- CompanyName_buick 0.4075	
- CompanyName_jaguar 0.2816 	

#### Notice that the constant term is not included as its p value is very high (0.228) and thus, is not significant. Which also means that if all these variables become 0 then the car price will also be 0 which makes sense as a car with 0 horsepower will not be a car so you can't sell it. So, constant term is 0.

#### Final Linear regression model equation is:-
    car_price_predicted =  0.7923 * horsepower - 0.0470 * carbody_hatchback + 0.2581 * enginelocation_rear + 0.0889 * enginetype_l + 0.2883 * CompanyName_bmw + 0.4075 * CompanyName_buick + 0.2816 * CompanyName_jaguar

    R-squared: 	0.872
    Adj. R-squared: 0.865
    Prob (F-statistic): 4.93e-57
    R-squared and Adjusted R-squared are pretty close to each other and more than 85% and Prob (F-statistic) is very low which shows that our model is a good fit.

### Residual Analysis for Training data and Regression line

```python
#Predicting the car price with our model and selected independent variables.
y_train_pred = lm_final.predict(sm.add_constant(X_train_model5))
```

```python
#calculating residuals
res = y_train - y_train_pred
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot(res, bins = 20)
# Plot heading 
fig.suptitle('Error Terms', fontsize = 20)                 
plt.xlabel('Errors', fontsize = 18)   
```

##### Analysis : Residuals are seemingly following a normal distribution with mean equal to zero 

```python
fig = plt.figure(figsize=(8, 5))
c = [i for i in range(1,144,1)]
fig.suptitle('Residual Plot', fontsize = 14)  
plt.scatter(c,res)
plt.hlines(xmin=0, xmax=144, y=0)                
# Plot heading 
plt.xlabel('Index', fontsize = 11) 
plt.ylabel('Residuals', fontsize = 11) 
plt.show()
```

##### Analysis :
    1. Homoscedasticity Assumption passed. Error terms seem to have constant variance.
    2. Residuals exhibits no pattern and are independent of each other.

### Extra Predictions - Using this model to predict price of 1 car
Here I have taken the horsepower of a car (63 units) and tried to predict the price of the car. In order to do it, i have performed the following steps:-
   1. Scaled (MinMax scaling) the horsepower using the same scaler object we fitted in our training dataset.
   2. Then I have created a DataFrame "mycar" which with all the significant variables values. For example, it's not a BMW so value would be 0.
   3. Then i have used the inverse_transformed function to get back the original car price back

```python
#Scaling horsepower. other values are 0 because they are not required.
hp_df = pd.DataFrame({'symboling':[0], 'carlength':[0], 'carwidth':[0], 'enginesize':[0], 'boreratio':[0], 'horsepower':[63], 'citympg':[0], 'price':[0]})
hp_scaled = scaler.transform(hp_df)

#Creating a dataframe with car features 
mycar = pd.DataFrame({'const':0, 'horsepower':hp_scaled[0][5], 'carbody_hatchback':[1], 'enginelocation_rear':[0], 
                         'enginetype_l':[1], 'CompanyName_bmw':[0], 'CompanyName_buick':[0], 'CompanyName_jaguar':[0]})
#My car price price prediction by using lm_final.predict. Saving the values in a new dataframe
mycar_pred = pd.DataFrame({'symboling':[0], 'carlength':[0], 'carwidth':[0], 'enginesize':[0], 'boreratio':[0], 
                           'horsepower':hp_scaled[0][5], 'citympg':[0], 'price':lm_final.predict(sm.add_constant(mycar))})
#inversing the transformation to get the actual price in USD
scaler.inverse_transform(mycar_pred)[0][7]
```
